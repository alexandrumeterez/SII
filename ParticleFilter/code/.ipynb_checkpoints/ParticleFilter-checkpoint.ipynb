```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
WIDTH = 400
HEIGHT = 400
N = 1000
```

```python
def uniform_particles(w, h, n):
    particles = np.empty((n, 3))
    particles[:, 0] = np.random.uniform(0, w, size=n)
    particles[:, 1] = np.random.uniform(0, h, size=n)
    particles[:, 2] = np.random.uniform(0, 2 * np.pi, size=n)
    return particles
```

```python
particles = uniform_particles(WIDTH, HEIGHT, 1000)
plt.figure()
plt.scatter(particles[:, 0], particles[:, 1], s=3)
plt.show()
```

```python
def predict(particles, u, std, dt):
    # u = [heading, direction]
    noise = np.random.normal(u, std)
    particles[:, 0] += np.cos(particles[:, 2]) * (u[1] * dt + noise[1])
    particles[:, 1] += np.sin(particles[:, 2]) * (u[1] * dt + noise[1])
    
    particles[:, 2] += u[0] + noise[0]
    particles[:, 2] %= (2 * np.pi)
```
